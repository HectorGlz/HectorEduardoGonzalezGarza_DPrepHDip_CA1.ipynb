```python
drop_columns = []

for column in df.columns:
    percentage = df[column].isnull().sum() / len(df)
    
    if(percentage >= 0.75):
        drop_columns = drop_columns + [column]
        
drop_columns
```

```python
df_dropped=df.drop(columns=drop_columns, axis=1)
df_dropped.head()
```
